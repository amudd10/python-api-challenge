```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keyss import g_key
```

```python
#loading csv file from weatherpy

load_file = ("../output_files/weather_py_dict.csv")

vacation_cities = pd.read_csv(load_file)
vacation_cities.head(20)
```

```python
max_humidiity = vacation_cities["Humidity"].max()
max_humidiity
humidity_rate = vacation_cities["Humidity"].astype(float)
```

```python
# Use the Lat and Lng as locations and Humidity as the weight
locations = vacation_cities[["Lat", "Lng"]]

fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights = humidity_rate,
                                 dissipating=False, max_intensity = max_humidiity,
                                 point_radius = 3)
fig.add_layer(heat_layer)
fig
```

```python
#converting temps to farenheit

vacation_cities['Max Temp'] = vacation_cities.apply(lambda x: (9/5)*x['Max Temp']+32,axis=1)
vacation_cities.head(20)
```

```python
# Narrow down: A max temperature lower than 80 degrees but higher than 70, Wind speed less than 10 mph, Zero cloudiness

vacation_cities_n = vacation_cities[(vacation_cities["Max Temp"] < 80) &
                                   (vacation_cities["Max Temp"] > 65) &
                                   (vacation_cities["Wind Speed"] < 10) &
                                    (vacation_cities["Cloudiness"] == 0)]
len(vacation_cities_n)
```

```python
# Perform API calls
# Plot the hotels on top of the humidity heatmap, with each pin containing the Hotel Name, City, and Country,
# set up lists to hold reponse info

hotel_df = pd.DataFrame(vacation_cities_n, columns = ["City", "Country", "Lat", "Lng"])

hotel_df["Hotel Name"] = ""
len(hotel_df)
```

```python
# set up lists to hold reponse info
name, city, country, lat, long = [], [], [], [], []

#data record count
record_count = 1

print("""Beginning Data Retrieval
---------------------------------""")

for i in range(len(vacation_cities_n)):
    
    latitude = str(vacation_cities_n.iloc[i,1])
    longitiude = str(vacation_cities_n.iloc[i,2])
    
    target_cooridnates = (f"{latitude},{longitiude}")
    target_search = 'Hotel'
    target_radius = 5000
    target_type = 'lodging'
    
    #paramters
    params = {
        'location': target_cooridnates,
#         'keyword': target_search,
        'radius': target_radius,
        'type': target_type,
        'key': g_key
    }
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    response = requests.get(base_url, params = params).json()
#     print(response)
    
    #Data Retrieval
    
    print('Searching for Hotels near: ' + str(vacation_cities_n.iloc[i, 0]) + ' | Record ' + str(record_count) + ' of Total '
         + str(len(vacation_cities_n)))
    
#     names = (response['results'][0]['name'])
#     cities = (vacation_cities_n.iloc[i,0])
#     countries = (vacation_cities_n.iloc[i,7])
#     lats = (response['results'][0]['geometry']['location']['lat'])
#     longs = (response['results'][0]['geometry']['location']['lng'])
    
    try:
        name.append(response['results'][0]['name'])
        
#         hotel_df.loc[i,"Hotel Name"] = response['results'][0]['name']
#         city.append(vacation_cities_n.iloc[i,0])
#         country.append(vacation_cities_n.iloc[i,7])
#         lat.append(response['results'][0]['geometry']['location']['lat'])
#         long.append(response['results'][0]['geometry']['location']['lng'])
    
    except IndexError:
        print('There seems to be no Hotels near Location... Skipping...')
        name.append ('')
    
    record_count +=1   

#print end of log
print("""---------------------------------
Data Retrieval Complete
---------------------------------""")   
```

```python
hotel_df["Hotel Name"] = name
hotel_df.head(25)
hotel_df.to_csv('Hotel CSV')
```
